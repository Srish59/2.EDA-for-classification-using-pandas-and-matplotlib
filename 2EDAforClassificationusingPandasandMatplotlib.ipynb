```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
df = pd.read_csv('car_prices--2q (1).csv')
df.head()
```

```python
df = df.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Style'], axis=1)
df.head(5)
```

```python
duplicate_rows_df = df[df.duplicated()]
print("Number of duplicate: ", duplicate_rows_df.shape)
```
